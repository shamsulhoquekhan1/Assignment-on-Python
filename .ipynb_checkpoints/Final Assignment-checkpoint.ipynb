## Task 1

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
plt.style.use('ggplot')
```

```python
# Importing the dataset from the given link

url = "https://data.ssb.no/api/v0/dataset/1088.csv?lang=en"

# We are creating a function change_cpi() that takes the given url link and a date in "YearMmonth", such as "2012M12", format
# and print the expected result. The data inside the link may change, but if it maintains the current formate, we will be
# able to extract data.

def change_cpi(link, date):
    ssb_df = pd.read_csv(link)
    ssb_df.columns = ['consumption_group', 'month', 'contents', 'cpi']
    s = str(12) + "-month rate (per cent)"
    ar = ssb_df[(ssb_df['month'] == date) & (ssb_df['contents'] == s)]['cpi'].values[0]
    
    return print('Change in CPI from December 2021 to December 2022 is: ' + str(ar) + ' %.')


change_cpi(url, "2022M12")
```

## Task 2

We will start by reading the datasets monthly_cpi.xlsx and monthly_cpi_ate.xlsx by using read_excel() function from pandas.

```python
monthly_cpi = pd.read_excel("data/monthly_cpi.xlsx")
monthly_cpi_ate = pd.read_excel("data/monthly_cpi_ate.xlsx")
```

It seems that we need to clean the datasets, because we need to cut out some of the unnecessary informations, give new understandable column names, and remove NaN values.

```python
# Cleaning monthly_cpi dataset

monthly_cpi.columns = ["year", "cpi"]

# We are dropping all the uncessary informations and NaA values

monthly_cpi.dropna(inplace = True)

# We will be plotting so making year column data time makes good sense.

monthly_cpi['year'] = pd.to_datetime(monthly_cpi['year'], format = "%YM%m")    



# Data from November 2020 to November 2022

cpi_recent = monthly_cpi.loc[(monthly_cpi['year'] >= '2019-11-01') & (monthly_cpi['year'] <= '2022-11-01')].copy()



# We will add another column 'cpi1' that represents percentage change over the 12 month period

cpi_recent['cpi1'] = ((cpi_recent['cpi'] - cpi_recent['cpi'].shift(12)) / cpi_recent['cpi'].shift(12)) * 100

cpi_recent.dropna(inplace = True)


# Cleaning monthly_cpi_ate dataset

monthly_cpi_ate.drop("05327: CPI-AT and CPI-ATE, by consumption group, month and contents", axis = 1, inplace = True)

monthly_cpi_ate.columns = ["year", "cpi_ate"]

monthly_cpi_ate.dropna(inplace = True)

# some values in "cpi_ate" columns are "..", so we will get rid of them.

monthly_cpi_ate = monthly_cpi_ate[monthly_cpi_ate['cpi_ate'] != '..'].copy() 


monthly_cpi_ate['year'] = pd.to_datetime(monthly_cpi_ate['year'], format = "%YM%m")

cpi_ate_recent = monthly_cpi_ate.loc[(monthly_cpi_ate['year'] >= '2019-11-01') & (monthly_cpi_ate['year'] <= '2022-11-01')].copy()

cpi_ate_recent['cpi_ate1'] = ((cpi_ate_recent['cpi_ate'] - cpi_ate_recent['cpi_ate'].shift(12)) / cpi_ate_recent['cpi_ate'].shift(12)) * 100

cpi_ate_recent.dropna(inplace = True)
```

```python
fig, ax = plt.subplots(figsize = (10, 8))
ax.set_facecolor('white')
ax.grid(color='gray', linestyle='-', linewidth=0.5)
ax.set_axisbelow(True)

ax.plot(cpi_recent['year'],
        cpi_recent['cpi1'],
        color = 'green',
        label = 'CPI')
ax.set_ylabel('CPI')    


ax.plot(cpi_ate_recent['year'],
        cpi_ate_recent['cpi_ate1'],
        color = 'black',
        label = 'CPI_ATE')
ax.set_ylabel('CPI_ATE')



ax.set_ylabel('Per cent', rotation = 1) 
ax.set_xlabel('Month')


#adjust position of both axis labels
ax.yaxis.set_label_coords(-.008, 1)
ax.xaxis.set_label_coords(.9, -.05)

plt.plot(cpi_recent['year'], cpi_recent['cpi1'], '-o')
plt.plot(cpi_ate_recent['year'], cpi_ate_recent['cpi_ate1'], '-o')



fig.suptitle('12 - month rate, CPI and CI-ATE. Per cent', fontsize = 16)
fig.tight_layout()
ax.legend() 
ax.legend(ncol=2, bbox_to_anchor=(0.5, -0.1))   # Changing the legend position
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Preparing dataset function 
def prepare_Dataset(url):
    df = pd.read_csv(url)
    
    #convert date format to datetime
    df["date"] = pd.to_datetime(df["month"], format="%YM%m")
    
    df = df.sort_values("month")
    df = df.set_index("month")

    
    # Calculate 12-month change in CPI for each group

    df = df[df["contents"] == "12-month rate (per cent)"]
    df["12-month change"] = df["03013: Consumer Price Index, by consumption group, month and contents"]
    df = df.drop("03013: Consumer Price Index, by consumption group, month and contents", axis = 1)
   

    df = df.drop(df[df["12-month change"] == ".."].index) 
    df["12-month change"] = df["12-month change"].astype(float)
    return df


#plotting cpi_change function
def plot_cpi_change(groups, start_date, end_date):
    df = prepare_Dataset("https://data.ssb.no/api/v0/dataset/1084.csv?lang=en")
    
    # Plot 12-month change in CPI for each group
    fig, ax = plt.subplots()
    
    for group in groups:
        try:
            group_df = df[df["consumption group"] == group]
        except:
            print("This group is not in the dataset: " + group_df)
        
        ax.plot(group_df['date'], group_df['12-month change'], label= group)
    ax.set_ylabel('12-month change in CPI (%)')
    ax.legend()
    plt.title("CPI by group of consumption goods")
    plt.show()
    

# Example usage:
from datetime import datetime
plot_cpi_change(["01 Food and non-alcoholic beverages", "03 Clothing and footwear"], "2021-01-01", "2022-12-01")

#Test exception handling

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#First Dataset
#Reading and preparing dataset - this import works as long as the dataset is in the same file as the jupyter notebook
df2 = pd.read_excel("data/monthly_earnings.xlsx") #######
df2 = df2[2:7].iloc[:,2:]

df2 = df2.T
df2 = df2.reset_index()
df2 = df2.drop("index", axis = 1)
df2 = df2.drop([1,2,3])
df2.columns = df2.iloc[0]
df2 = df2.iloc[1:]

# df2["Year"] = df2.iloc[:,0]

# df2 = df2.astype(int)
df2["Year"] = pd.to_numeric(df2.iloc[:,0], errors='coerce').astype(int)


print(df2)


# Create a new dataframe for storing the yearly growth by sector
growth_df = pd.DataFrame(columns=["Year", "Sector", "Yearly Growth (%)"])

# Calculate the growth for each sector and add it to the new dataframe
sectors = ["Sum all sectors",  "Private sector and public enterprises",  "Local government", "Central government"]
for sector in sectors:
    for year in range(2016, 2022):
        growth = (df2.loc[df2["Year"] == year, sector].iloc[0] - df2.loc[df2["Year"] == year-1, sector].iloc[0]) / df2.loc[df2["Year"] == year-1, sector].iloc[0] * 100
#         growth_df = growth_df.append({"Year": year, "Sector": sector, "Yearly Growth (%)": growth}, ignore_index=True)
        growth_df = pd.concat([growth_df, pd.DataFrame({"Year": [year], "Sector": [sector], "Yearly Growth (%)": [growth]})], ignore_index=True)  

#---------------------------------------------------------------------------------------------
#Second Dataset
#Reading and preparing dataset - this import works as long as the dataset is in the same file as the jupyter notebook
df_cpi = pd.read_excel("data/monthly_cpi.xlsx")  ########
df_cpi = df_cpi[2:529]
df_cpi = df_cpi.reset_index()

df_cpi = df_cpi.drop("index", axis = 1)
df_cpi.columns = ['Year', 'CPI']
df_cpi["Year"] = pd.to_datetime(df_cpi["Year"], format="%YM%m")
df_cpi["CPI"].astype(float)

df_cpi = df_cpi[pd.DatetimeIndex(df_cpi['Year']).month == 11].copy()
df_cpi = df_cpi[pd.DatetimeIndex(df_cpi['Year']).year > 2014].copy()
df_cpi = df_cpi.reset_index()
df_cpi = df_cpi.drop("index", axis = 1)



#First option to calculate differences
df_cpi["CPI %"] = df_cpi["CPI"].diff()

#New column for alternative way of calculating CPI
df_cpi["CPI2 %"] = None

#Iterating through the date and calculating the Inflation compared to the previous year
for i in range(1, len(df_cpi)):
    prev_year_cpi = df_cpi.loc[i-1, "CPI"]
    current_year_cpi = df_cpi.loc[i, "CPI"]
    inflation = (current_year_cpi - prev_year_cpi) / prev_year_cpi * 100
    df_cpi.loc[i, "CPI2 %"] = inflation
df_cpi = df_cpi.drop(0)
df_cpi = df_cpi.reset_index()
df_cpi = df_cpi.drop("index", axis = 1)
df_cpi["Year"] = pd.DatetimeIndex(df_cpi['Year']).year.astype(str) 

# Creating a bar blot like above
plt.figure(figsize=(10,10))
growth_df.pivot("Year", "Sector", "Yearly Growth (%)").plot(kind='bar')


# Adding CPI Data to the line graph
plt.plot(df_cpi['Year'], df_cpi['CPI2 %'], '-o')

# Adding labels for the x and y axis
plt.xlabel("Year")
plt.ylabel("CPI / Percent Growth")

#Adding legend
plt.legend(["CPI", "Percent Growth"])
plt.title("Wage growth versus inflation")
#Showing plot
plt.show()

 
```

```python
#first install package via: conda install -c https://conda.anaconda.org/anaconda pandas-datareader
import pandas_datareader as pdr
from pandas_datareader import wb

#Getting data for special indicators from worldbank API
names = ["Inflation, consumer prices (annual %)", "Broad money (% of GDP)", "Domestic credit to private sector (% of GDP)", "Market capitalization of listed domestic companies (% of GDP)", "GDP growth (annual %)", "Gross savings (% of GDP)", "Imports of goods and services (% of GDP)"]
indicatorids = {"FP.CPI.TOTL.ZG","FM.LBL.BMNY.GD.ZS", "FS.AST.PRVT.GD.ZS", "CM.MKT.LCAP.GD.ZS", "NY.GDP.MKTP.KD.ZG", "NY.GNS.ICTR.ZS", "NE.IMP.GNFS.ZS"}

worldbank_df = wb.download(indicator = indicatorids, country='all', start='2019', end='2019')
worldbank_df.columns = names

# Scatter plots for different explainatory variables
explanatory_vars = names[1:]
for var in explanatory_vars:
    plt.scatter(worldbank_df['Inflation, consumer prices (annual %)'], worldbank_df[var])
    plt.xlabel('Inflation, consumer prices (annual %)')
    plt.ylabel(var)
    plt.title(var + " and " + "Inflation", fontsize=10, color = "b")
    plt.show()

```
